### Menghapus unnamed columns

```python
def hapus_unnamed_kolom(data):
    data_bersih = data.loc[:, ~data.columns.str.contains('^Unnamed')]
    return data_bersih
```

```python
# data_banjir_kecamatan = hapus_unnamed_kolom(data_banjir_kecamatan)
data_banjir_kelurahan = hapus_unnamed_kolom(data_banjir_kelurahan)
```
